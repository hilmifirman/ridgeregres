Ridge Regression

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
diabetes = pd.read_csv("diabetes.csv")
```

```python
diabetes
```

```python
train, test = train_test_split(diabetes, test_size=0.2, random_state=1)
```

```python
predictors = ["Age", "Insulin"]
target = "Outcome"
```

```python
X = train[predictors].copy()
y = train[[target]].copy()
```

```python
X
```

```python
y
```

```python
x_mean = X.mean()
x_std = X.std()
```

```python
x_mean
```

```python
x_std
```

```python
X = (X - x_mean) / x_std
```

```python
X
```

```python
X["intercept"] = 1
```

```python
X = X[["intercept"] + predictors]
```

```python
X.describe()
```

```python
X.T
```

```python
alpha = 2
I = np.identity(X.shape[1])
penalty = alpha * I
```

```python
penalty
```

```python
penalty[0][0] = 0
```

```python
penalty
```

```python
B = np.linalg.inv(X.T @ X + penalty) @ X.T @ y
```
